```python
import pandas as pd
```

```python
data_genre = pd.read_csv('id_genres_tf-idf.tsv', sep='\t')
data_genre.head()
```

```python
print(data_genre.shape)
```

```python
# Dictionary to store the top 100 songs (video_id, score) per genre
genre_to_top100 = {}

for genre in data_genre.columns:
    if genre == 'id':
        continue
    # Get top 100 non-zero entries for the genre
    top_songs = data_genre[['id', genre]].nlargest(100, genre)
    # Filter out rows with 0 score
    top_songs = top_songs[top_songs[genre] > 0]
    # Convert to list of tuples: (video_id, score)
    genre_to_top100[genre] = list(top_songs.itertuples(index=False, name=None))
```

```python
# Optional: example output format
for genre, songs in list(genre_to_top100.items()):  # show first 5 genres
    print(f"{genre}: {songs[:5]}")
    break
```

```python
from collections import defaultdict

def assign_unique_top_k_songs_per_genre(genre_to_top100, k):
    assigned_videos = set()
    final_assignments = defaultdict(list)

    genres = list(genre_to_top100.keys())

    for genre in genres:
        top_candidates = genre_to_top100[genre]
        assigned = []
        fallback_index = k  # start checking fallback from k+1-th candidate

        for vid, score in top_candidates:
            if vid not in assigned_videos:
                assigned.append((vid, score))
                assigned_videos.add(vid)
                if len(assigned) == k:
                    break

        if len(assigned) == k:
            final_assignments[genre] = assigned
            continue

        # Conflict resolution
        temp_set = set(vid for vid, _ in assigned)

        for vid, score in top_candidates:
            if len(assigned) == k:
                break
            if vid in temp_set:
                continue
            if vid in assigned_videos:
                # find current owner
                for g, vids in final_assignments.items():
                    if vid in [v for v, _ in vids]:
                        other_genre = g
                        break

                other_candidates = genre_to_top100[other_genre]
                this_fallback = next((s for v, s in top_candidates if v not in assigned_videos and v != vid), 0)

                other_fallback = other_candidates[k][1] if len(other_candidates) > k else 0

                if other_fallback > this_fallback:
                  # Other genre has better fallback → reassign to current genre
                  final_assignments[other_genre] = [entry for entry in final_assignments[other_genre] if entry[0] != vid]
                  assigned_videos.remove(vid)

                  assigned.append((vid, score))
                  assigned_videos.add(vid)

                  # Assign fallback to other genre
                  for next_vid, next_score in other_candidates:
                      if next_vid not in assigned_videos:
                          final_assignments[other_genre].append((next_vid, next_score))
                          assigned_videos.add(next_vid)
                          break
                else:
                  continue  # Other genre has worse fallback → let them keep the song

            else:
                assigned.append((vid, score))
                assigned_videos.add(vid)

        final_assignments[genre] = assigned[:k]

    return list(final_assignments.items())
```

```python
import pandas as pd

stats = pd.DataFrame(columns=[
    'unique songs I assign to genre',
    'genre I can cover totally',
    'genres I cannot fully cover',
    'total genres'
])

# Check how many genres can be assigned at least i music videos
for i in range(1, 16):
    genre_assignment = assign_unique_top_k_songs_per_genre(genre_to_top100, i)

    total_covered = 0
    total_not_covered = 0

    for genre, assigned_songs in genre_assignment:
        if len(assigned_songs) == i:
            total_covered += 1
        else:
            total_not_covered += 1

    row = {
        'unique songs I assign to genre': i,
        'genre I can cover totally': total_covered,
        'genres I cannot fully cover': total_not_covered,
        'total genres': total_covered + total_not_covered
    }
    stats = pd.concat([stats, pd.DataFrame([row])], ignore_index=True)
stats
```
