```python
# Dictionary to store all non-zero songs (video_id, score) per genre
genre_sorted = {}

for genre in data_genre.columns:
    if genre == 'id':
        continue
    # Filter out rows with 0 score
    non_zero_songs = data_genre[data_genre[genre] > 0][['id', genre]]
    # Sort by score descending
    non_zero_songs = non_zero_songs.sort_values(by=genre, ascending=False)
    # Convert to list of tuples: (video_id, score)
    genre_sorted[genre] = list(non_zero_songs.itertuples(index=False, name=None))
```

```python
max_song = None
max_genre = None
max_score = 0

for genre, songs in list(genre_sorted.items()):
    for song in songs:
        if max_score < song[1]:
            max_genre = genre
            max_song = song[0]
            max_score = song[1]
print(max_genre, max_song, max_score)
```

```python
from collections import defaultdict

def assign_k_per_genre(genre_sorted, k):
    assignments = {}
    next_idx = {}

    for genre, items in genre_sorted.items():
        if len(items) < k:
            continue
        assignments[genre] = items[:k].copy()
        next_idx[genre] = k  # next candidate position (0-based)

    def current_conflicts(assignments):
        holders = defaultdict(list)  # video_id -> [genres]
        for g, vids in assignments.items():
            for vid, _ in vids:
                holders[vid].append(g)
        return {vid: gs for vid, gs in holders.items() if len(gs) > 1}

    def fallback_score(genre):
        items = genre_sorted[genre]
        idx = next_idx[genre]
        if idx < len(items):
            return items[idx][1]
        return float("-inf")

    def tiebreak_key(genre, contested_video_id):
        # Higher current score for this contested video should win;
        # if still tie, lexicographically smaller genre name wins.
        cur_score = next(s for (v, s) in assignments[genre] if v == contested_video_id)
        return (-cur_score, genre)

    while True:
        conflicts = current_conflicts(assignments)
        if not conflicts:
            break

        # Handle each contested video independently this round
        for video_id, genres in conflicts.items():
            # Choose the winner: genre with the *smallest* fallback score
            # (smaller fallback is worse alternative → gets to keep original)
            # If tie on fallback, use tiebreaker that prefers the genre with higher current score for this video, then name.
            fallbacks = [(fallback_score(g), g) for g in genres]
            min_fallback = min(fs for fs, _ in fallbacks)
            candidates = [g for fs, g in fallbacks if fs == min_fallback]
            if len(candidates) == 1:
                winner = candidates[0]
            else:
                # deterministic tie-break
                winner = sorted(candidates, key=lambda g: tiebreak_key(g, video_id))[0]

            losers = [g for g in genres if g != winner]

            # For each loser: remove the contested video, then try to fill up to k with next unique candidates
            for g in losers:
                # Remove the contested video from this genre’s assignment
                assignments[g] = [(v, s) for (v, s) in assignments[g] if v != video_id]

            # Build a global set of currently assigned videos (after removals)
            assigned_videos = set()
            for vids in assignments.values():
                assigned_videos.update(v for v, _ in vids)

            # Now refill losers up to k with next available unique videos
            for g in losers:
                items = genre_sorted[g]
                i = next_idx[g]
                while len(assignments[g]) < min(k, len(items)) and i < len(items):
                    vid, sc = items[i]
                    i += 1
                    if vid in assigned_videos:
                        continue
                    assignments[g].append((vid, sc))
                    assigned_videos.add(vid)
                next_idx[g] = i  # advance pointer

        # loop again to see if new conflicts arose

    # Remove genres with less than k assigned videos
    assignments = {g: vids for g, vids in assignments.items() if len(vids) == k}

    # Prepare unmet report (now optional, since we only keep genres with k videos)
    unmet = {g: max(0, k - len(assignments[g])) for g in assignments}
    return assignments, unmet
```

```python
import json
import pickle
import pandas as pd
from urllib.parse import urlparse, parse_qs
from typing import Dict, List, Tuple

# --------- helpers ----------
def extract_youtube_id(url: str) -> str | None:
    """
    Extract VIDEO_ID from common YouTube URL forms:
    - https://www.youtube.com/watch?v=VIDEO_ID&...
    - https://youtu.be/VIDEO_ID
    - https://www.youtube.com/embed/VIDEO_ID
    - https://www.youtube.com/shorts/VIDEO_ID
    Returns None if not found.
    """
    if not isinstance(url, str):
        return None

    u = urlparse(url)
    host = (u.netloc or "").lower()
    path = u.path or ""

    # youtu.be short links
    if host.endswith("youtu.be"):
        vid = path.lstrip("/").split("/")[0]
        return vid or None

    # youtube.com (incl. music.youtube.com)
    if host.endswith("youtube.com"):
        if path.startswith("/watch"):
            q = parse_qs(u.query or "")
            return (q.get("v") or [None])[0]
        if path.startswith("/embed/"):
            return path.split("/embed/")[1].split("/")[0] or None
        if path.startswith("/shorts/"):
            return path.split("/shorts/")[1].split("/")[0] or None

    return None


def dedup_keep_max(items: List[Tuple[str, float]]) -> List[Tuple[str, float]]:
    """
    Deduplicate (video_id, score) by video_id, keeping the entry with the highest score.
    Returns sorted desc by score.
    """
    best = {}
    for vid, sc in items:
        if vid is None:
            continue
        if vid not in best or sc > best[vid]:
            best[vid] = sc
    return sorted(best.items(), key=lambda x: x[1], reverse=True)


# --------- load mapping df ----------
# Your example:
# ID                         url
# NDroPROgWm3jBxjH           https://www.youtube.com/watch?v=gPm2s6JORc4
# ...
with open("../data/music4all_to_youtube.json", "r", encoding="utf-8") as f:
    df = pd.DataFrame(json.load(f))

# Extract the true YouTube IDs
df["video_id"] = df["url"].apply(extract_youtube_id)

# Build mapping: dataset ID -> true YouTube video ID
id_to_youtube: Dict[str, str] = dict(zip(df["ID"], df["video_id"]))

missing_count = df["video_id"].isna().sum()
if missing_count:
    print(f"Warning: {missing_count} URLs had no extractable YouTube video ID. They will be dropped.")

# --------- load assignments and remap ----------
with open("../data/assignments_top50.pkl", "rb") as f:
    assignments: Dict[str, List[Tuple[str, float]]] = pickle.load(f)
    # assignments[genre] == [(dataset_id, score), ...]

mapped: Dict[str, List[Tuple[str, float]]] = {}
dropped_total = 0

for genre, items in assignments.items():
    mapped_items: List[Tuple[str, float]] = []
    for ds_id, score in items:
        yt_id = id_to_youtube.get(ds_id)
        if yt_id:
            mapped_items.append((yt_id, score))
        else:
            dropped_total += 1
    # optional: deduplicate within genre
    mapped[genre] = dedup_keep_max(mapped_items)

print(f"Done. Dropped {dropped_total} items with no valid YouTube ID across all genres.")

# --------- save new assignments ----------
with open("../data/assignments_top50_youtube.pkl", "wb") as f:
    pickle.dump(mapped, f)

print("Saved remapped assignments to ../data/assignments_top50_youtube.pkl")
```
