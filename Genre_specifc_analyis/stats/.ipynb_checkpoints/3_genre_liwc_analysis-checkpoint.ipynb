#### compare Balance between emo_pos and emo neg
    - a lot of positive comments and litle negative will reach a high store

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import patheffects as pe  # outlined text keeps labels readable

# ============== CONFIG ==============
MIN_N = 500
FORCE_ALIGN_LABELS = True          # True = always place genre labels at the anchor column
LABEL_X_ANCHOR_DIFF = 5         # anchor column (x) for genre labels (in "% of words")
BASE_MIN_RIGHT = 0.35              # minimal right x-limit for small ranges
VALUE_GAP_FRAC = 0.01              # gap after bar for numeric value (fraction of x-span)
LABEL_EXTRA_FRAC = 0.03            # extra gap if we must push label after bar
# ===================================

# --- Prep data ---
for col in ['emo_pos', 'emo_neg']:
    data[col] = pd.to_numeric(data[col], errors='coerce')

agg_emotion = (
    data.dropna(subset=['genre'])
        .groupby('genre', dropna=True)
        .agg(
            emo_pos=('emo_pos', 'mean'),
            emo_neg=('emo_neg', 'mean'),
            n=('commentid', 'size')
        )
        .query('n >= @MIN_N')
)

# Balance metric
agg_emotion['diff_pos_minus_neg'] = agg_emotion['emo_pos'] - agg_emotion['emo_neg']

# Select extremes
ranked   = agg_emotion.sort_values('diff_pos_minus_neg', ascending=False)
top5     = ranked.head(5)
bottom5  = ranked.tail(5)

# Keep order: top first, then bottom (without duplicates)
order = list(top5.index) + [g for g in bottom5.index if g not in top5.index]
extremes = agg_emotion.loc[order].copy()

# Colors
colors = ["steelblue" if g in top5.index else "lightcoral" for g in extremes.index]

# --- Plot (labels outside, aligned column) ---
fig, ax = plt.subplots(figsize=(12, 7))

y = np.arange(len(extremes))
vals = extremes['diff_pos_minus_neg'].values.astype(float)

bars = ax.barh(
    y=y,
    width=vals,
    left=0.0,                 # bars extend left or right from zero
    color=colors,
    edgecolor='black'
)

# Highest at top
ax.invert_yaxis()

# Build genre labels (we draw them ourselves; hide y-ticks)
genre_labels = [f"{g} (n={int(extremes.loc[g,'n'])})" for g in extremes.index]
ax.set_yticks(y)
ax.set_yticklabels([""] * len(y))

# ----- X limits with room for negatives, positives, and labels -----
vmin = float(np.nanmin(vals)) if len(vals) else 0.0
vmax = float(np.nanmax(vals)) if len(vals) else 0.0

# start with tight bounds plus headroom
left  = vmin * 1.20 if vmin < 0 else 0.0
right = vmax * 1.20 if vmax > 0 else BASE_MIN_RIGHT

# ensure a sensible minimum width
right = max(right, BASE_MIN_RIGHT)

# ensure we can see the anchor column + its label width
max_len = max((len(s) for s in genre_labels), default=0)
span_guess = max(right - left, 1.0)
char_unit = max(0.01 * span_guess, 0.02) * 0.5  # data-units per character (rough)
label_room = max_len * char_unit
right = max(right, LABEL_X_ANCHOR_DIFF + label_room + 0.05)

# ensure room for numeric text after the longest (absolute) bar
right = max(right, max(abs(vmin), abs(vmax)) + 0.08 * max(right, 1.0))

ax.set_xlim(left, right)

# ----- Texts -----
span = right - left
pad_num = max(0.005, VALUE_GAP_FRAC * span)   # gap for numeric value
pad_lab = max(0.02,  LABEL_EXTRA_FRAC * span) # extra if we push label after bar
outline = [pe.withStroke(linewidth=3, foreground="white")]  # to keep readable over bars

for i, (v, label) in enumerate(zip(vals, genre_labels)):
    v = float(v)

    # numeric value at the bar tip (left or right)
    if v >= 0:
        ax.text(v + pad_num, i, f"{v:.3f}", va='center', ha='left',
                fontsize=11, fontweight='bold')
    else:
        ax.text(v - pad_num, i, f"{v:.3f}", va='center', ha='right',
                fontsize=11, fontweight='bold')

    # genre label column behavior
    if FORCE_ALIGN_LABELS:
        # Always place at fixed anchor
        x_label = LABEL_X_ANCHOR_DIFF
        ax.text(x_label, i, label, va='center', ha='left',
                fontsize=11, path_effects=outline)
    else:
        # Align at anchor unless the bar crosses it; then push after bar
        threshold_left  = LABEL_X_ANCHOR_DIFF + 0.02 * span if v < 0 else LABEL_X_ANCHOR_DIFF - 0.02 * span
        crosses_anchor = (v >= 0 and v >= threshold_left) or (v < 0 and v <= LABEL_X_ANCHOR_DIFF)
        if crosses_anchor:
            x_label = (v + pad_num + pad_lab) if v >= 0 else (v - pad_num - pad_lab)
            ha = 'left' if v >= 0 else 'right'
            ax.text(x_label, i, label, va='center', ha=ha, fontsize=11)
        else:
            ax.text(LABEL_X_ANCHOR_DIFF, i, label, va='center', ha='left',
                    fontsize=11, path_effects=outline)

# Title / axes
ax.set_title("Top 5 and Bottom 5 Genres by LIWC Emotion Balance (emo_pos âˆ’ emo_neg)",
             fontsize=18, fontweight='bold', pad=12)
ax.set_xlabel("Mean(emo_pos) âˆ’ Mean(emo_neg) (% of words)", fontsize=14, fontweight='bold')
ax.set_ylabel("Genres", fontsize=14, fontweight='bold')

# Zero line
ax.axvline(0, color='black', linewidth=0.8)

# Clean spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Legend
legend_handles = [
    plt.Rectangle((0,0),1,1,color="steelblue",  edgecolor="black", label="Top 5 Genres"),
    plt.Rectangle((0,0),1,1,color="lightcoral", edgecolor="black", label="Bottom 5 Genres")
]
ax.legend(handles=legend_handles, loc="lower right")

plt.tight_layout()
os.makedirs('plots/liwc', exist_ok=True)
plt.savefig('plots/liwc/top_bottom_emopos_emoneg.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import patheffects as pe  # for outlined text

MIN_N = 500
features = [
    "Tone", "Health", "Religious", "Illness", "Death", "Friend", "Social",
    "Family", "Swear", "Anger", "Anxiety", "Sadness", "Politeness",
    "Conflict", "Politics", "Ethnicity", "Culture"
]

# Always align genre labels at the feature-specific anchor?
FORCE_ALIGN_LABELS = True  # <- set False to allow "push after bar" behavior

# --- Per-feature label anchors (x-position where genre labels should align) ---
LABEL_X_ANCHORS = {
    "tone": 100.0,
    "health": 1.4,
    "religious": 2.0,
    "illness": 0.8,
    "death": 3.0,
    "friend": 1.0,
    "social": 20.0,      # Social has large values; column at 3.5
    "family": 1.4,
    "swear": 4.5,
    "anger": 2.0,
    "anxiety": 2.0,
    "sadness": 2.0,
    "politeness": 2.0,
    "conflict": 2.0,
    "politics": 2.0,
    "ethnicity": 1.0,
    "culture": 1.75,
}
DEFAULT_LABEL_X_ANCHOR = 2.0  # fallback

# Lowercase all columns (matches LIWC names provided)
data.columns = [c.lower() for c in data.columns]

# Output folder
os.makedirs("plots/liwc", exist_ok=True)

def compute_xlim_for_labels(values, label_texts, label_anchor, base_min_right=1.75):
    """
    X-limits:
      - start at 0
      - include all bars
      - ensure >= base_min_right
      - ensure label column (anchor) + label width are visible
      - ensure room for numeric values after the longest bar
    """
    vmax = float(np.nanmax(values)) if len(values) else 0.0
    right = max(vmax * 1.20 if vmax > 0 else 1.0, base_min_right)

    max_len = max((len(s) for s in label_texts), default=0)
    # approx data-units per character, scaled with current span
    char_unit = max(0.01 * max(right, 1.0), 0.02) * 0.5
    label_room = max_len * char_unit

    # Make sure we can see the anchor + label
    right = max(right, label_anchor + label_room + 0.25)
    # Room for numeric value after the longest bar
    right = max(right, vmax + 0.08 * max(right, 1.0))

    return 0.0, right

for feat in features:
    col = feat.lower()
    if col not in data.columns:
        print(f"Skipping {feat} â€” not in dataset")
        continue

    # Ensure numeric
    data[col] = pd.to_numeric(data[col], errors="coerce")

    # Aggregate by genre
    agg = (
        data.dropna(subset=["genre"])
            .groupby("genre", dropna=True)
            .agg(mean_feat=(col, "mean"),
                 n=("commentid", "size"))
            .query("n >= @MIN_N")
            .sort_values("mean_feat", ascending=False)
    )

    if agg.empty:
        print(f"Skipping {feat} â€” no valid data after filtering")
        continue

    # Top & Bottom 5
    top5 = agg.head(5)
    bottom5 = agg.tail(5)
    extremes = pd.concat([top5, bottom5]).drop_duplicates()
    order = list(top5.index) + [g for g in bottom5.index if g not in top5.index]
    extremes = extremes.loc[order].copy()

    # Colors: top=blue, bottom=coral
    colors = [("steelblue" if g in top5.index else "lightcoral") for g in extremes.index]

    # Precompute arrays
    vals = extremes["mean_feat"].values.astype(float)
    y = np.arange(len(extremes))
    genre_labels = [f"{g} (n={int(n)})" for g, n in zip(extremes.index, extremes["n"].values)]

    # Feature-specific label anchor (fallback to default)
    label_anchor = LABEL_X_ANCHORS.get(col, DEFAULT_LABEL_X_ANCHOR)

    # --- Plot ---
    fig, ax = plt.subplots(figsize=(13, 7))
    ax.barh(y, vals, color=colors, edgecolor="black")
    ax.invert_yaxis()

    # X-limits with room for labels
    xmin, xmax = compute_xlim_for_labels(vals, genre_labels, label_anchor, base_min_right=1.75)
    ax.set_xlim(xmin, xmax)

    span = xmax - xmin
    pad_num = max(0.005, 0.01 * span)   # gap for numeric values after bar
    pad_gen = max(0.02, 0.03 * span)    # extra if we place label after bar (when not forcing alignment)

    # Text outline to keep labels readable over bars
    outline = [pe.withStroke(linewidth=3, foreground="white")]

    for i, (v, label) in enumerate(zip(vals, genre_labels)):
        v = max(0.0, v)

        # numeric value at bar end
        ax.text(v + pad_num, i, f"{v:.3f}",
                va="center", ha="left", fontsize=11, fontweight="bold")

        if FORCE_ALIGN_LABELS:
            # Always at the anchor (strict vertical column)
            x_label = label_anchor
            ax.text(x_label, i, label, va="center", ha="left",
                    fontsize=11, path_effects=outline)
        else:
            # Only align if the bar doesn't reach the anchor; otherwise push after bar
            threshold = label_anchor - 0.02 * span
            if v < threshold:
                x_label = label_anchor
                ax.text(x_label, i, label, va="center", ha="left",
                        fontsize=11, path_effects=outline)
            else:
                x_label = v + pad_num + pad_gen
                ax.text(x_label, i, label, va="center", ha="left", fontsize=11)

    # Titles / axes
    ax.set_title(f"Top 5 and Bottom 5 Genres by LIWC {feat} (English-only)",
                 fontsize=18, fontweight="bold", pad=12)
    ax.set_xlabel(f"Average LIWC '{feat}' Score (% of words)",
                  fontsize=14, fontweight="bold")
    ax.set_ylabel("Genres", fontsize=14, fontweight="bold")

    # Hide default y tick labels (we render our own)
    ax.set_yticks(y)
    ax.set_yticklabels([""] * len(y))

    # Zero line
    ax.axvline(0, color="black", linewidth=0.8)

    # Clean spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Legend
    ax.legend(handles=[
        plt.Rectangle((0, 0), 1, 1, color="steelblue",  edgecolor="black", label="Top 5 Genres"),
        plt.Rectangle((0, 0), 1, 1, color="lightcoral", edgecolor="black", label="Bottom 5 Genres")
    ], loc="lower right")

    plt.tight_layout()
    out = f"plots/liwc/{col}.png"
    plt.savefig(out, dpi=150, bbox_inches="tight")
    plt.close()

    print(f"Saved plot for {feat} -> {out}")
```

```python
data = pd.read_csv('../data/final_dataset_english_only.csv')

numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()

exclude_cols = [
    'viewCount', 'likeCount_y', 'commentCount', 'Unnamed: 0', 'likeCount_x',
    'Dic', 'Linguistic', 'function', 'Analytic', 'WC', 'WPS',
    'pronoun','verb', 'det', 'adj', 'space', 'ppron', 'BigWords', 'prep'
]

feature_cols = [c for c in numeric_cols if c not in exclude_cols]

genre_means = (
    data.groupby('genre')[feature_cols]
        .mean()
        .dropna(axis=1, how='all')
)

feature_variance = genre_means.var().sort_values(ascending=False)

top_features = feature_variance.head(10).index.tolist()

print("Top 10 features with highest variance across genres:\n")
print(feature_variance.head(10))

plt.figure(figsize=(10,5))
bars = plt.bar(range(len(top_features)), feature_variance.head(10).values,
               color="steelblue", edgecolor="black")

for i,v in enumerate(feature_variance.head(10).values):
    plt.text(i, v, f"{v:.2f}", ha='center', va='bottom', fontsize=9)

plt.xticks(range(len(top_features)), top_features, rotation=45, ha="right")
plt.title("Top 10 LIWC/VADER Features with Most Variance Across Genres", fontsize=13)
plt.ylabel("Variance across Genres")
plt.tight_layout()
plt.show()
```

```python
top_features = [
    "Tone", "Clout", "Authentic", "Emoji", "Cognition",
    "Social", "Perception", "Affect", "tone_pos", "allure"
]

for col in top_features:
    data[col] = pd.to_numeric(data[col], errors="coerce")

genre_means = (
    data.dropna(subset=['genre'])
        .groupby('genre')[top_features]
        .mean()
)

MIN_N = 500
genre_counts = data['genre'].value_counts()
valid_genres = genre_counts[genre_counts >= MIN_N].index
genre_means = genre_means.loc[valid_genres]

genre_means_norm = (genre_means - genre_means.mean()) / genre_means.std()

plt.figure(figsize=(14, 10))
sns.heatmap(
    genre_means_norm,
    cmap="coolwarm",
    center=0,
    cbar_kws={'label': 'z-score (normalized feature values)'},
    yticklabels=True,
    xticklabels=top_features
)

plt.title("Genre Clustering by Emotion & Semantics (Top-Variance Features)", fontsize=14)
plt.ylabel("Genre")
plt.xlabel("Feature")
plt.tight_layout()
plt.savefig("plots/liwc/genre_clustering_top10_semantic.png")
plt.show()
```

```python
available_features = [f for f in top_features if f in genre_means.columns]

long_df = (
    genre_means_norm.reset_index()
        .melt(id_vars="genre", var_name="feature", value_name="z_score")
)

long_df = long_df[long_df["feature"].isin(available_features)]

long_df["raw_value"] = long_df.apply(
    lambda r: genre_means.loc[r["genre"], r["feature"]], axis=1
)
long_df["global_avg"] = long_df["feature"].apply(
    lambda f: genre_means[f].mean()
)

long_df["abs_dev"] = long_df["z_score"].abs()

top10_devs = long_df.sort_values("abs_dev", ascending=False).head(10)

print("Top 10 (genre, feature) deviations from average:\n")
top10_devs
```
